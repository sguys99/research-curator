## 3. API í´ë¼ì´ì–¸íŠ¸ í…ŒìŠ¤íŠ¸

```python
from app.frontend.utils.api_client import APIClient

# Create API client instance
api_client = APIClient(base_url="http://localhost:8000")

print("ğŸŒ API Client Initialized")
print(f"  Base URL: {api_client.base_url}")
print(f"  Timeout: {api_client.timeout}s")

# List all methods
methods = [name for name, obj in inspect.getmembers(api_client) 
           if inspect.ismethod(obj) and not name.startswith('_')]

print(f"\nğŸ“ API Client Methods ({len(methods)}):")
print("=" * 50)

# Group by category
categories = {
    "Authentication": ["request_magic_link", "verify_magic_link"],
    "User Management": ["get_current_user", "get_user_preferences", "update_user_preferences"],
    "Articles": ["get_articles", "get_article"],
    "Semantic Search": ["search_articles", "find_similar_articles"],
    "Digests": ["get_user_digests", "get_latest_digest", "send_test_digest"],
    "Feedback": ["submit_feedback", "get_user_feedback"],
    "LLM": ["chat_completion"],
}

for category, expected_methods in categories.items():
    print(f"\n{category}:")
    for method in expected_methods:
        exists = method in methods
        status = "âœ…" if exists else "âŒ"
        print(f"  {status} {method}()")
```
